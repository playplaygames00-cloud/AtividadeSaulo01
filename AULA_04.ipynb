**Desafio 3: Triângulos**

Peça os 3 lados de um triângulo. O programa deverá informar se os valores formam um triângulo. Caso formem, diga o tipo do triângulo:

Equilátero: 3 lados iguais;
Isósceles: 2 lados iguais;
Escaleno: 3 lados diferentes. Dica: Para formar um triângulo, a soma de quaisquer dois lados precisa ser maior que o terceiro lado.

```python
letra = (input ("Digite uma letra e te diremos se ela é vogal ou consoante:").lower())

if letra in ("a","e","i","o","u"):
  print ("A sua letra é uma VOGAL!")
else :
  print ("A sua letra é uma CONSOANTE!")
```

```python
salario = float(input("Qual o seu salário atual: "))

if salario <= 1500:
    print (f"Seu aumento é de 15% e seu novo salário é de {round (salario*1.15, 2)}")
else  :
    print (f"Seu aumento é de 10% e seu novo salário é de {round (salario*1.10, 2)}")
```

```python
ladoA = float(input("Digite o tamanho do 1º lado: "))
ladoB = float(input("Digite o tamanho do 2º lado: "))
ladoC = float(input("Digite o tamanho do 3º lado:"))

if  ladoA + ladoB > ladoC and ladoA + ladoC > ladoB and ladoB + ladoC > ladoA:

  if ladoA == ladoB == ladoC:
    print("Você consegue formar um Triângulo equilátero!")

  elif ladoA == ladoB or ladoA == ladoC or ladoB == ladoC :
    print("Você consegue formar um Triângulo isósceles")

  else :
    print("Você consegue formar um Triângulo escaleno")

else :
  print("Isso não é um Triângulo!")
```
