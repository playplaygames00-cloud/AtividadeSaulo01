**Desafio 1: Acima da Média**

Você recebeu uma lista com as idades de um grupo de pessoas. Calcule a média dessas idades (usando sum() e len()) e depis, use um for para imprimir somente as idades que são MAIORES que a média.

```python
idades = [25, 45, 18, 30, 22, 60, 52]

media_idades = (sum(idades))/(len(idades))

for idade in idades:
  if idade > media_idades:
    print (idade)
```
