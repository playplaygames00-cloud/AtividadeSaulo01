```python
from IPython.core.display import clear_output
print("===BEM VINDO A SUA CALCULADOR IMC===")
print("Para começarmos digite sua altura em Metros:")
print("Por exemplo: '1.78' ou '1.50'")
altura = float(input())
clear_output (True)

print("Digite seu peso em KG:")
print("Por exemplo: '78' ou '75.5'")
peso = float(input())
clear_output (True)

imc = peso / (altura * altura)

if imc < 18.5:
    result = "Você está abaixo do peso ideal"

elif imc >= 18.5 and imc <= 24.9:
    result = "Você está no peso ideal"

elif imc >= 25 and imc <= 29.9:
    result = "Você está com sobrepeso"

elif imc >= 30 and imc <= 39.9:
    result = "Você está com obesidade"

elif imc >= 40:
    result = "Você está com obesidade grave"

print("CALCULANDO...")
input("Aperte [ENTER] para continuar...")

clear_output (True)

print(f"Seu peso é de {round(peso, 2)} Kg ")
print(f"Sua altura é de {round(altura,3)} Metros")
print(f"Seu IMC é de {round(imc,2)}")
print(result)
```
