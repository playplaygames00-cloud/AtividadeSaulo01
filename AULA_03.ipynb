# OBJETIVO DA ATIVIDADE

Um parque de diversões tem um brinquedo radical. Para entrar nele, a pessoa precisa:

Ter pelo menos 1.50m de altura (altura >= 1.50)
Ter 12 anos ou mais (idade >= 12)
Porém, se a pessoa estiver acompanhada dos pais, basta ter 1.30m de altura (independente da idade).

Escreva as perguntas usando input() e faça a árvore de decisão if/elif/else com and/or para liberar ou barrar a pessoa.

```python
print ("Vamos verificar se você pode brincar neste brinquedo")

idade = float (input ("Digite a sua idade: "))
altura = float (input ("Digite a sua altura em metros: [ex: 1.30m] "))

print ("Você está com um responsável maior de idade? ")
resposta_user = int (input ("1-) Sim \n2-) Não"))


if (resposta_user) == 1:
   verifica_pais = (True)
else :
   verifica_pais = (False)


if altura < 1.30 :
  print ("Você não tem altura para brincar neste brinquedo! ")
elif altura >= 1.30 and verifica_pais == True :
  print ("Você pode brincar neste brinquedo!")
elif altura >= 1.50 and idade >= 12:
  print ("Você pode brincar neste brinquedo!")
else :
  print ("Você não pode brincar neste brinquedo!")
```
