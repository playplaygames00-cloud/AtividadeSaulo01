**Desafio 2: Maior de Idade**

Crie uma função chamada verifica_maioridade que recebe uma idade. A função não deve ter print. Ela deve retornar a string "Maior de Idade" ou "Menor de Idade".

```python
def ehpar (n):
  if n % 2 == 0:
    return True
  else :
    return False


n = int(input("Digite seu número e diremos se ele é par ou ímpar  "))

if ehpar(n) == True:
  print(f"Seu número {n} é par")
else :
  print(f"Seu número {n} é ímpar")


```
